```python
import pandas as pd
import plotly.offline as pyo
import plotly.graph_objects as go
```

```python
df = pd.read_pickle('../Data/data.pkl')
```

```python
df.head()
```

```python
data = [go.Scatter(
        x = df['Verwendungszweck'],
        y = df['Zusageobligo'],
        mode = 'lines',
        name = 'Zusageobligo'
        )] 

layout = go.Layout(title = 'Zusageobligo pro Verwendungszweck')

fig = go.Figure(data=data, layout=layout)

pyo.plot(fig, filename='Zusageobligo pro Verwendungszweck.html')
```
