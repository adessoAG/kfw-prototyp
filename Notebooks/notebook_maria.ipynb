```python
data = data.fillna(0)
```

```python
sums = data.iloc[: ,1:188].apply(sum, axis = 0)
```

```python
sums_df = pd. DataFrame(sums).T
```

```python
sums_df.columns
```

```python
sgds = []

for col in sums_df.columns:
    if 'SDG' in str(col):
        sgds.append(col)
```

```python
sgds
```

```python
sgd = sums_df[sgds].values
```

```python
ratios = []
for i in sgd.tolist():
    for j in i:
        ratios.append(j/sgd.sum()*100)
```

```python
ratios
```
